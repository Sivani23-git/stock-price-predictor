```python
!pip install yfinance scikit-learn matplotlib pandas numpy tensorflow

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

ticker = 'AAPL'
df = yf.download(ticker, start='2015-01-01', end='2024-12-31')


df_lr = df[['Close']].copy()
df_lr['Target'] = df_lr['Close'].shift(-1)

df_lr.dropna(inplace=True)

X = df_lr[['Close']]
y = df_lr['Target']

X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X, y, test_size=0.2, random_state=0)

lr_model = LinearRegression()
lr_model.fit(X_train_lr, y_train_lr)

pred_lr = lr_model.predict(X_test_lr)

plt.figure(figsize=(10,5))
plt.title('Linear Regression - Closing Price Prediction')
plt.plot(y_test_lr.values, label='Actual')
plt.plot(pred_lr, label='Predicted')
plt.legend()
plt.show()


scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['Close']])

sequence_len = 60
X_lstm = []
y_lstm = []

for i in range(sequence_len, len(scaled_data)):
    X_lstm.append(scaled_data[i-sequence_len:i, 0])
    y_lstm.append(scaled_data[i, 0])

X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)
X_train, X_test = X_lstm[:int(len(X_lstm)*0.8)], X_lstm[int(len(X_lstm)*0.8):]
y_train, y_test = y_lstm[:int(len(y_lstm)*0.8)], y_lstm[int(len(y_lstm)*0.8):]

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=10, batch_size=32)

pred_lstm = model.predict(X_test)
pred_lstm = scaler.inverse_transform(pred_lstm.reshape(-1, 1))
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))


plt.figure(figsize=(10,5))
plt.title('LSTM - Closing Price Prediction')
plt.plot(actual_prices, label='Actual')
plt.plot(pred_lstm, label='Predicted')
plt.legend()
plt.show()
```
